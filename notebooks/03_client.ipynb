```python
import requests
import json

urls = {
    'Logistic Regression': 'http://127.0.0.1:8001/predict',
    'SVM': 'http://127.0.0.1:8002/predict',
    'Decision Tree': 'http://127.0.0.1:8003/predict',
    'KNN': 'http://127.0.0.1:8004/predict'
}
```

```python
penguin1 = {
    "island": "Torgersen",
    "bill_length_mm": 39.1,
    "bill_depth_mm": 18.7,
    "flipper_length_mm": 181.0,
    "body_mass_g": 3750.0,
    "sex": "Male"
}

print("=" * 60)
print("PREDICCIÓN PARA PINGÜINO 1")
print("=" * 60)
print(f"Características: {json.dumps(penguin1, indent=2)}")
print("\nEste pingüino tiene características típicas de un Adelie:")
print("- Pico corto y profundo")
print("- Aleta mediana")
print("- Peso moderado")
print("\n" + "=" * 60)

for model_name, url in urls.items():
    try:
        response = requests.post(url, json=penguin1)
        result = response.json()
        print(f"\n{model_name}:")
        print(f"  Predicción: {result['predicted_species']}")
        if result.get('probabilities'):
            print(f"  Probabilidades:")
            for species, prob in result['probabilities'].items():
                if prob is not None:
                    print(f"    - {species}: {prob:.4f} ({prob*100:.2f}%)")
    except Exception as e:
        print(f"\n{model_name}: ERROR - {e}")
```
