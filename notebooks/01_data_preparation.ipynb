```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = sns.load_dataset("penguins")
print(f"Total de registros: {len(df)}")
df.head()
```

```python
print(df.info())
print("\nValores nulos por columna:")
print(df.isnull().sum())
```

```python
df_clean = df.dropna()
print(f"Registros después de limpiar: {len(df_clean)}")
print(f"Registros eliminados: {len(df) - len(df_clean)}")
```

```python
print(df_clean['species'].value_counts())
df_clean['species'].value_counts().plot(kind='bar')
plt.title('Distribución de especies')
plt.show()
```

```python
df_clean.to_csv('../datasets/penguins_clean.csv', index=False)
print("Dataset limpio guardado en datasets/penguins_clean.csv")
```
